```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sin
import scipy.signal
```

```python
""" 3.a """
def sampled_sin (signal, Ts, tmin, tmax):
    n = np.arange((int)(tmin / Ts), (tmax / Ts))
    y_out = map(lambda m : signal(m*Ts), n)
    
    return n, list(y_out)
```

```python
""" ΑΜ: 03115082 άρα 0+8+2=10 και 1+0=1 οπότε fm = 1 kHz """
fm = 1*10**3 #Hz
Tm = 1.0 / fm #sec
A = 1 #Volt
z = lambda t : A*sin(2*pi*fm*t)

fs2 = 130 * fm
Ts2 = 1.0 / fs2
mu = 0.5
m = lambda t : sin(2*pi*35*t)

s = lambda t : z(t) * (1 + mu * m(t))

tmin, tmax = 0, 4*Tm #4 periods
n, y = sampled_sin (s, Ts2, tmin, tmax)

plt.plot(n*Ts2, y, linewidth=0.5)
plt.title('DSBxAmulated signal with mu=0.5')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (V)')
plt.show()
```
